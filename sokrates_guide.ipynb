# Sokrates

Sokrates is a tool for generating documentation from source code. It supports multiple programming languages and can be easily integrated into your development workflow.
I have tried to use it to analyze zeeguu-api, following the guide on https://www.sokrates.dev.

First, i downloaded 'sokrates_LATEST.jar' file from https://www.sokrates.dev.

Then, from the root of the zeeguu-api directory, I ran the following command to generate documentation for the zeeguu-api project:

```bash
java -jar <sokrates-folder>/sokrates-LATEST.jar extractGitHistory
java -jar <sokrates-folder>/sokrates-LATEST.jar init
java -jar <sokrates-folder>/sokrates-LATEST.jar generateReports

open _sokrates/reports/html/index.html
```

The last line opens the generated documentation in a web browser. Here, I could explore the documentation for the zeeguu-api project, including class diagrams, method summaries, and other relevant information extracted from the source code.

(I later moved the generated '_sokrates' directory into this repo) 

# Observations

Generally, the analysis looks at level one components, but this is not really the most interesting in this case. I would much rather go depper inside the zeeguu-folder.
To see more, i changed componentsFolderDepth to 2 in the logicalDecompositions object in the sokrates configuration file (_sokrates/config.json) and ran the generateReports command again. 

## Overview
The repo is 9 years old (there will be a lot of history). The language is mainly python. Database is MySQL
82% of the main code has been touched in the last year and 50% of the main code is new (created within the last year). 

In the diagram in the bottom we can choose contributors and we will see one orange unit in a mostly blue landscape: the user.py file. This means that lot of different contributors (11-25) have made changes to this file, and also that most files are usually not touched by a lot of different developers (the responsibility/knowledge might thereby not be very distributed) 


## Goals & Controls
It warns us that there is a high amount of duplication, 5 very large files (longer than 1000 LOC) and 3 very complex units (with more than 25 decision points)

We can easily locate the big files by looking at the overview page and choosing bubble diagram in the bottom colored by sizes.  We can also get them listed on the File Size page. 
It might be interesting to investigate if these could be split into smaller units. 
The File Size page also shows Files With Most Units and Files With Long Lines which are both helpfull to determine the type of problem.
